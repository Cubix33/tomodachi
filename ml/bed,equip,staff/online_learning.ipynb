```python
#FINAL

import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential, model_from_json
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
import os
import pickle

# Load dataset
df = pd.read_csv("bed_data.csv")

# Convert Date column to datetime and create Is_Weekend feature
if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"])
    df["Is_Weekend"] = df["Date"].dt.dayofweek.isin([5, 6]).astype(int)  # 1 for Sat/Sun, 0 otherwise
    df.set_index("Date", inplace=True)

# Select features & target
features = [
    "Total_Admissions_Today",
    "Total_Discharges_Today",
    "Avg_LOS",
    "Avg_Age_Admissions_Today",
    "Total_Beds_Occupied_Today",
    "Is_Weekend"
]
target = "Total_Beds_Required_Tomorrow"

# Normalize features & target
scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[features + [target]] = scaler.fit_transform(df[features + [target]])

# Save scaler for later use
scaler_path = "scaler.pkl"
with open(scaler_path, "wb") as f:
    pickle.dump(scaler, f)

# Train-Test Split (80% Train, 20% Test)
train_size = int(len(df_scaled) * 0.8)
train, test = df_scaled.iloc[:train_size], df_scaled.iloc[train_size:]

# Prepare training data (reshape for LSTM)
X_train = train[features].values.reshape(-1, 1, len(features))
y_train = train[target].values
X_test = test[features].values.reshape(-1, 1, len(features))
y_test = test[target].values

# Define LSTM Model
def build_lstm_model():
    model = Sequential([
        LSTM(64, input_shape=(1, len(features))),  # Single timestep input
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(loss="mse", optimizer="adam")
    return model

# Paths for saving model
model_path = "xyz.pkl"

# Load or Train Model
if os.path.exists(model_path):
    with open(model_path, "rb") as f:
        model_json, weights = pickle.load(f)
    model = model_from_json(model_json)
    model.set_weights(weights)
    model.compile(loss="mse", optimizer="adam")
    print("✅ Loaded existing model.")
else:
    model = build_lstm_model()
    model.fit(X_train, y_train, epochs=100, batch_size=16, validation_data=(X_test, y_test), verbose=1)

    # Save model as JSON + Weights using Pickle
    model_json = model.to_json()
    weights = model.get_weights()
    with open(model_path, "wb") as f:
        pickle.dump((model_json, weights), f)
    print("✅ Trained and saved new model.")

# Predict next day's beds
latest_data = df_scaled.iloc[-1][features].values.reshape(1, 1, len(features))  # Single row input
predicted_scaled = model.predict(latest_data)

# Reverse scale the prediction
latest_data_unscaled = scaler.inverse_transform(
    np.append(latest_data.reshape(1, -1), [[0]], axis=1)  # Append a dummy column for inverse scaling
)
predicted_beds_unscaled = scaler.inverse_transform(
    np.append(latest_data_unscaled[:, :-1], predicted_scaled.reshape(-1, 1), axis=1)
)[:, -1][0]

print(f"🛏️ Predicted Beds for Tomorrow: {predicted_beds_unscaled:.2f}")
```

```python
#FINAL

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load dataset and prepare test data
df = pd.read_csv("bed_data.csv")

# Convert Date column to datetime and create Is_Weekend feature
if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"])
    df["Is_Weekend"] = df["Date"].dt.dayofweek.isin([5, 6]).astype(int)
    df.set_index("Date", inplace=True)

# Define features and target
features = [
    "Total_Admissions_Today",
    "Total_Discharges_Today",
    "Avg_LOS",
    "Avg_Age_Admissions_Today",
    "Total_Beds_Occupied_Today",
    "Is_Weekend"
]
target = "Total_Beds_Required_Tomorrow"

# Train-test split
train_size = int(len(df) * 0.8)
train_data = df.iloc[:train_size]
test_data = df.iloc[train_size:]

# Load the scaler and model
with open("scaler.pkl", "rb") as f:
    scaler = pickle.load(f)

with open("xyz.pkl", "rb") as f:
    model_json, weights = pickle.load(f)
model = model_from_json(model_json)
model.set_weights(weights)
model.compile(loss="mse", optimizer="adam")

# Get actual test values
y_test_actual = test_data[target].values

# Make predictions on test data
predictions = []
for i in range(len(test_data)):
    row = test_data.iloc[i:i+1][features]
    
    # Add dummy target column for scaler
    row_with_dummy = row.copy()
    row_with_dummy[target] = 0  # Dummy value
    
    # Scale the input
    row_scaled = scaler.transform(row_with_dummy)
    
    # Remove the target column for model input
    row_scaled_features = row_scaled[:, :-1]
    
    # Reshape for LSTM (samples, time steps, features)
    row_reshaped = row_scaled_features.reshape(1, 1, len(features))
    
    # Make prediction
    pred_scaled = model.predict(row_reshaped, verbose=0)
    
    # Prepare for inverse scaling
    row_with_pred = np.zeros((1, len(features) + 1))
    row_with_pred[:, :-1] = row_scaled_features
    row_with_pred[:, -1] = pred_scaled.flatten()
    
    # Inverse transform to get actual values
    row_with_pred_actual = scaler.inverse_transform(row_with_pred)
    
    # Get the predicted value
    pred_actual = row_with_pred_actual[0, -1]
    predictions.append(pred_actual)

y_pred_actual = np.array(predictions)

# Get latest prediction (for tomorrow)
latest_row = df.iloc[-1:][features]
latest_row_with_dummy = latest_row.copy()
latest_row_with_dummy[target] = 0

latest_row_scaled = scaler.transform(latest_row_with_dummy)
latest_row_features = latest_row_scaled[:, :-1]
latest_row_reshaped = latest_row_features.reshape(1, 1, len(features))

latest_pred_scaled = model.predict(latest_row_reshaped, verbose=0)

latest_row_with_pred = np.zeros((1, len(features) + 1))
latest_row_with_pred[:, :-1] = latest_row_features
latest_row_with_pred[:, -1] = latest_pred_scaled.flatten()

latest_row_with_pred_actual = scaler.inverse_transform(latest_row_with_pred)
predicted_beds = latest_row_with_pred_actual[0, -1]

# Check if we have actual data for "tomorrow"
actual_beds = None
if len(df) > train_size + len(test_data):
    actual_beds = df.iloc[train_size + len(test_data)][target]

# Plot historical values
plt.figure(figsize=(12, 6))
plt.plot(range(len(y_test_actual)), y_test_actual, label="Actual Beds Required", marker="o")
plt.plot(range(len(y_pred_actual)), y_pred_actual, label="Predicted Beds", marker="x")

# Add the latest prediction vs actual (if available)
plt.scatter(len(y_test_actual), predicted_beds, color='blue', label="Predicted Beds (Tomorrow)", marker="D", s=100)
if actual_beds is not None:
    plt.scatter(len(y_test_actual), actual_beds, color='red', label="Actual Beds (Tomorrow)", marker="s", s=100)

plt.xlabel("Days")
plt.ylabel("Beds Required")
plt.legend()
plt.title("Actual vs Predicted Bed Requirements")
plt.grid(True, alpha=0.3)
plt.savefig("bed_prediction_plot.png", dpi=300, bbox_inches="tight")
plt.show()

print(f"🛏️ Predicted Beds for Tomorrow: {predicted_beds:.2f}")
if actual_beds is not None:
    print(f"📊 Actual Beds for Tomorrow: {actual_beds:.2f}")
    print(f"📉 Error: {abs(predicted_beds - actual_beds):.2f} beds ({abs(predicted_beds - actual_beds)/actual_beds*100:.2f}%)")
```
